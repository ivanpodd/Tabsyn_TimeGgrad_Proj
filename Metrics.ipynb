```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metrics_1e3 = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_orig/metrcis.csv')
metrics_3e3 = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_lr_3e3/metrcis.csv')
metrics_6e3 = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_lr_6e3//metrcis.csv')
```

```python
metrics_1e3_long = pd.melt(metrics_1e3, ['epoch'])
metrics_3e3_long = pd.melt(metrics_3e3, ['epoch'])
metrics_6e3_long = pd.melt(metrics_6e3, ['epoch'])
```

```python
metrics_1e3_long
```

```python
sns.lineplot(data=metrics_1e3_long, x='epoch', y='value', hue='variable')
```

```python
sns.lineplot(data=metrics_3e3_long, x='epoch', y='value', hue='variable')
```

```python
sns.set(rc = {'figure.figsize':(20, 15)})
sns.set_style("whitegrid")
sns.lineplot(data=metrics_6e3_long, x='epoch', y='value', hue='variable')
```

```python
metrics_1e3_nouser = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_split_nouser_1e3/metrcis.csv')
metrics_5e4_nouser = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_split_nouser_5e4/metrcis.csv')
metrics_1e4_nouser = pd.read_csv('tabsyn/tabsyn/vae/ckpt/transactions_dt_split_nouser_1e4/metrcis.csv')
```

```python
metrics_1e3_long_nores = pd.melt(metrics_1e3_nouser, ['epoch'])
metrics_5e4_long_nores = pd.melt(metrics_5e4_nouser, ['epoch'])
metrics_1e4_long_nores = pd.melt(metrics_1e4_nouser, ['epoch'])
```

```python
sns.set(rc = {'figure.figsize':(20, 15)})
sns.set_style("whitegrid")
sns.lineplot(data=metrics_1e3_long_nores, x='epoch', y='value', hue='variable')
```

```python
sns.set(rc = {'figure.figsize':(20, 15)})
sns.set_style("whitegrid")
sns.lineplot(data=metrics_5e4_long_nores, x='epoch', y='value', hue='variable')
```

```python
sns.set(rc = {'figure.figsize':(20, 15)})
sns.set_style("whitegrid")
sns.lineplot(data=metrics_1e4_long_nores, x='epoch', y='value', hue='variable')
```

```python
sns.set(rc = {'figure.figsize':(12, 8)})
sns.set_style("whitegrid")
sns.lineplot(data=metrics_1e3_long_nores[metrics_1e3_long_nores.variable == 'kl_loss'], x='epoch', y='value')
plt.xlabel("", fontsize=18, weight = 'bold')
plt.ylabel("Loss", fontsize=18, weight = 'bold')
plt.title('KL divergence loss', fontsize=18, weight = 'bold')
plt.xticks(fontsize=12, weight = 'normal')
plt.yticks(fontsize=12, weight = 'normal')
plt.savefig('kl_loss.png')
plt.ylim(0, 5.5)
plt.show()
```
